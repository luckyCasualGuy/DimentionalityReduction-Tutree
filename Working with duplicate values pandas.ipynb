# Finding Duplicate values

```python
df = pd.read_csv('dataset/car.csv')
```

```python
df
```

## In order to see the count of Say Car_Names repeated you can do something like this
<a id="Actual_Count"></a>

```python
NAME = "Car_Name"

n = df.groupby(by=[NAME]).count().iloc[:, 0:1]
n.columns.values[0] = 'Count'

# Making a function out of it for quick use
def duplicate_count(df):
    n = df.groupby(by=[NAME]).count().iloc[:, 0:1]
    n.columns.values[0] = 'Count'
    return n

n
```

## If you just wont to know if particular columns of your dataset has duplicate values you can do something like this

```python
NAME = "Car_Name"

m = df[NAME].duplicated().any()
n = df[NAME].is_unique

f"Value of m: {m} ## Value of n: {n}"
```

## Say Now we want to remove every duplicate car names from Car_Name column but keeping the first occurence

```python
NAME = "Car_Name"

n = df.drop_duplicates(subset=[NAME])

duplicate_count(n)
```

### You can Now see that our new dataframe contains only one occurence of that car and rest  were deleted
